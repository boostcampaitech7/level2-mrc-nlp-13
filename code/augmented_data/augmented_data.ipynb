```python
import json

def load_and_preview_data(file_path):
    # JSON 파일 읽기
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    
    # 데이터에서 'data' 키의 내용 출력
    if 'data' in data:
        print(f"Contents of 'data' in {file_path}:")
        for i, item in enumerate(data['data']):
            if i < 5:  # 처음 5개 항목 출력
                print(item)
            else:
                break
    else:
        print(f"'data' key not found in {file_path}")
```

```python
train_file = data_path + 'train.json'
load_and_preview_data(train_file)
```

```python
dev_file = data_path + 'dev.json'
load_and_preview_data(dev_file)
```
